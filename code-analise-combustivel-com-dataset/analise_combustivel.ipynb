`````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()

pd.set_option('display.max_rows', 1000)
# Objetivo: Configura o pandas para exibir at√© 1000 linhas de um DataFrame ao imprimi-lo no console.
# Uso: √ötil quando voc√™ est√° analisando datasets grandes e quer visualizar mais linhas sem que o pandas oculte parte delas com "..." (retic√™ncias).
`````

`````python
df = pd.read_csv('dataset/precos_semestrais_combustivel_automotivo_2024.02.csv', encoding='utf-8-sig', sep=';')
`````

`````python
# Pagando os 5 primeiros dados
df.head()
`````

`````python
# Pegando os 5 √∫ltimos dados
df.tail()
`````

`````python
df.info()
`````

`````python
# Excluindo as colunas que n√£o ser√£o necess√°rias nesse projeto
excluir = ['CNPJ da Revenda', 'CNPJ da Revenda', 'Nome da Rua', 'Numero Rua', 'Complemento', 'Bairro', 'Cep', 'Valor de Compra']

df.drop(excluir, axis=1, inplace=True)
`````

`````python
df.head()
`````

`````python
# Criando uma fun√ß√£o que transformar√° as v√≠rgulas dos valores em pontos. 
# E na mesma fun√ß√£o transformar√° os valores em float.
def removeVirgula(valor):
    try:
        valor = str(valor).replace(',', '.')
        return float(valor)
    except:
        return np.NaN # Retorna Valor Ausente

# Aplicando a fun√ß√£o a cada valor da coluna 'Valor de Venda' no DataFrame df
df['Valor de Venda'] = df['Valor de Venda'].map(removeVirgula)
`````

`````python
df.info()
`````

### ü§î **Qual Estado possui a gasolina mais cara?**

`````python
dfGasolina = df[df['Produto'] == 'GASOLINA']

dfGasolina
`````

`````python
dfGasolina.groupby(['Estado - Sigla']).max().sort_values('Valor de Venda', ascending=False)
`````

`````python
# Gr√°fico de Barras: Pre√ßo M√°ximo da Gasolina Registrado por Estado (2¬∫ Sem 2024)
preco_max_gasolina = dfGasolina.groupby(['Estado - Sigla'])['Valor de Venda'].max().sort_values(ascending=False)
plt.figure(figsize=(12, 7))
plt.bar(preco_max_gasolina.index, preco_max_gasolina.values)
plt.title('Pre√ßo M√°ximo da Gasolina Registrado por Estado (2¬∫ Sem 2024)')
plt.xlabel('Estado')
plt.ylabel('Pre√ßo M√°ximo (R$/litro)')
plt.xticks(rotation=90) # Rotaciona para caber todos os estados
plt.ylim(bottom=preco_max_gasolina.min() * 0.95) # Ajusta limite inferior do eixo Y
plt.tight_layout()
plt.show()
`````

---

### ü§î **Qual Estado possui a gasolina mais barata?**

`````python
dfGasolina.groupby(['Estado - Sigla']).min().sort_values('Valor de Venda')
`````

`````python
# Gr√°fico de Barras: Pre√ßo M√≠nimo da Gasolina Registrado por Estado (2¬∫ Sem 2024)
preco_min_gasolina = dfGasolina.groupby(['Estado - Sigla'])['Valor de Venda'].min().sort_values()
plt.figure(figsize=(12, 7))
plt.bar(preco_min_gasolina.index, preco_min_gasolina.values)
plt.title('Pre√ßo M√≠nimo da Gasolina Registrado por Estado (2¬∫ Sem 2024)')
plt.xlabel('Estado')
plt.ylabel('Pre√ßo M√≠nimo (R$/litro)')
plt.xticks(rotation=90)
plt.ylim(bottom=preco_min_gasolina.min() * 0.9) # Ajusta limite inferior
plt.tight_layout()
plt.show()
`````

### ü§î **Qual o pre√ßo m√©dio de gasolina por Estado?**

`````python
dfGasolina.groupby(['Estado - Sigla']).mean(numeric_only=True)
`````

üïµÔ∏è‚Äç‚ôÇÔ∏è __OBSERVA√á√ÉO:__

````python
.mean(numeric_only=True)
````

- ``numeric_only=True:`` ignora colunas n√£o num√©ricas (como strings ou datas), considerando apenas colunas num√©ricas no c√°lculo da m√©dia.
- Isso evita erros caso o DataFrame tenha colunas de texto ou outros tipos de dados.

`````python
dfGasolina.groupby(['Estado - Sigla']).mean(numeric_only=True).sort_values('Valor de Venda', ascending=False)
`````

`````python
# Gr√°fico de Barras: Pre√ßo M√©dio da Gasolina por Estado (2¬∫ Sem 2024)
preco_medio_gasolina = dfGasolina.groupby(['Estado - Sigla'])['Valor de Venda'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 7))
plt.bar(preco_medio_gasolina.index, preco_medio_gasolina.values)
plt.title('Pre√ßo M√©dio da Gasolina por Estado (2¬∫ Sem 2024)')
plt.xlabel('Estado')
plt.ylabel('Pre√ßo M√©dio (R$/litro)')
plt.xticks(rotation=90)
plt.ylim(bottom=preco_medio_gasolina.min() * 0.98) # Ajusta limite inferior
plt.axhline(dfGasolina['Valor de Venda'].mean(), color='r', linestyle='--', label=f'M√©dia Brasil ({dfGasolina["Valor de Venda"].mean():.2f})') # Linha da m√©dia nacional
plt.legend()
plt.tight_layout()
plt.show()
`````
