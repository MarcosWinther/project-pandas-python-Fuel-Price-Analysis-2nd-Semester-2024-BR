# ⛽📊 Exploração de Dados de Combustíveis - 2º Semestre 2024 (Pandas + NumPy)

- **Dataset:** ``2° Sem 2024 - Combustíveis Automotivos`` disponível no [Gov - Portal de Dados Abertos](https://dados.gov.br/dados/conjuntos-dados/serie-historica-de-precos-de-combustiveis-e-de-glp)

`````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()

pd.set_option('display.max_rows', 1000)
# Objetivo: Configura o pandas para exibir até 1000 linhas de um DataFrame ao imprimi-lo no console.
# Uso: Útil quando você está analisando datasets grandes e quer visualizar mais linhas sem que o pandas oculte parte delas com "..." (reticências).
`````

`````python
df = pd.read_csv('dataset/precos_semestrais_combustivel_automotivo_2024.02.csv', encoding='utf-8-sig', sep=';')
`````

`````python
# Pagando os 5 primeiros dados
df.head()
`````

`````python
# Pegando os 5 últimos dados
df.tail()
`````

`````python
df.info()
`````

`````python
# Excluindo as colunas que não serão necessárias nesse projeto
excluir = ['CNPJ da Revenda', 'CNPJ da Revenda', 'Nome da Rua', 'Numero Rua', 'Complemento', 'Bairro', 'Cep', 'Valor de Compra']

df.drop(excluir, axis=1, inplace=True)
`````

`````python
df.head()
`````

`````python
# Criando uma função que transformará as vírgulas dos valores em pontos. 
# E na mesma função transformará os valores em float.
def removeVirgula(valor):
    try:
        valor = str(valor).replace(',', '.')
        return float(valor)
    except:
        return np.NaN # Retorna Valor Ausente

# Aplicando a função a cada valor da coluna 'Valor de Venda' no DataFrame df
df['Valor de Venda'] = df['Valor de Venda'].map(removeVirgula)
`````

`````python
df.info()
`````

### 🤔 **Qual Estado possui a gasolina mais cara?**

`````python
dfGasolina = df[df['Produto'] == 'GASOLINA']

dfGasolina
`````

`````python
dfGasolina.groupby(['Estado - Sigla']).max().sort_values('Valor de Venda', ascending=False)
`````

`````python
# Gráfico de Barras: Preço Máximo da Gasolina Registrado por Estado (2º Sem 2024)
preco_max_gasolina = dfGasolina.groupby(['Estado - Sigla'])['Valor de Venda'].max().sort_values(ascending=False)
plt.figure(figsize=(12, 7))
plt.bar(preco_max_gasolina.index, preco_max_gasolina.values)
plt.title('Preço Máximo da Gasolina Registrado por Estado (2º Sem 2024)')
plt.xlabel('Estado')
plt.ylabel('Preço Máximo (R$/litro)')
plt.xticks(rotation=90) # Rotaciona para caber todos os estados
plt.ylim(bottom=preco_max_gasolina.min() * 0.95) # Ajusta limite inferior do eixo Y
plt.tight_layout()
plt.show()
`````

### 🤔 **Qual o preço médio de gasolina por Estado?**

`````python
dfGasolina.groupby(['Estado - Sigla']).mean(numeric_only=True)
`````

🕵️‍♂️ __OBSERVAÇÃO:__

````python
.mean(numeric_only=True)
````

- ``numeric_only=True:`` ignora colunas não numéricas (como strings ou datas), considerando apenas colunas numéricas no cálculo da média.
- Isso evita erros caso o DataFrame tenha colunas de texto ou outros tipos de dados.

`````python
dfGasolina.groupby(['Estado - Sigla']).mean(numeric_only=True).sort_values('Valor de Venda', ascending=False)
`````

`````python
# Gráfico de Barras: Preço Médio da Gasolina por Estado (2º Sem 2024)
preco_medio_gasolina = dfGasolina.groupby(['Estado - Sigla'])['Valor de Venda'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 7))
plt.bar(preco_medio_gasolina.index, preco_medio_gasolina.values)
plt.title('Preço Médio da Gasolina por Estado (2º Sem 2024)')
plt.xlabel('Estado')
plt.ylabel('Preço Médio (R$/litro)')
plt.xticks(rotation=90)
plt.ylim(bottom=preco_medio_gasolina.min() * 0.98) # Ajusta limite inferior
plt.axhline(dfGasolina['Valor de Venda'].mean(), color='r', linestyle='--', label=f'Média Brasil ({dfGasolina["Valor de Venda"].mean():.2f})') # Linha da média nacional
plt.legend()
plt.tight_layout()
plt.show()
`````
